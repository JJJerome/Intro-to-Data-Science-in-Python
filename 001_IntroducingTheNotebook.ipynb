# Markdown cells

In addition to code cells, there are also markdown cells for entering prose, taking notes, annotating code and figures, etc. 

Markdown cells also support Latex commands for math notation. For example, typing `$f(x) = \alpha x^2$` produces

$f(x) = \alpha x^2$

Moreover, `$f(x) = \alpha x^2$` was produced by surrounding the whole expression with backticks, i.e. \``like this `\`.

Headings are produced with \# symbols. The greater the number of \#s, the more subordinate the heading.

# Heading
written as 
`# Heading`
## Sub-heading
written as 
`## Sub-heading`
### etc..
written as 
`### etc..`

## Code highlighting
We can even write python code in markdown mode for display with appropriate highlighting and formatting. For example:

```python
def myfunc(x):
    y = x.dosomething()
    return y
```

## Keyboard shortcuts

The shortcuts improve flow. The ones used most by me are `enter` to edit a cell, `esc` to enter command mode, once in command mode `M` for switching the cell to markdown, `Y` for switching it to code. `shift-enter` runs a cell and moves to the next or creates one if none is there. `control-enter` runs a cell and leaves it selected.

### Notebook magic
Using the percent sign `%` is not standard python syntax. In the notebook, this is known as 'magic', andin this case it imports numpy as np, matplotlib.pyplot as plt, ands sets us up to print figures directly in the notebook. For example, 

````python
x = np.linspace(-100, 100, 1000)
plt.figure(figsize=(20,10))
plt.plot(x,  sin(x)*sin(.1*x)*cos(.02*x), 'k', lw=4)
plt.plot(x, -sin(x)*sin(.1*x)*cos(.02*x), 'k', lw=4)
plt.show()
````
