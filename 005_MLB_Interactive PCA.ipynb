```python
import pandas as pd
from bokeh.plotting import figure, output_file, output_notebook, show
from bokeh.models.sources import ColumnDataSource
from bokeh.models import HoverTool
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# output plot inline
output_notebook()

# tools documentation found here: http://bokeh.pydata.org/en/0.11.1/docs/user_guide/tools.html
```

```python
df = pd.read_csv('data/MLB_2.csv').dropna()
```

```python

df.head()
```

```python
plt.scatter(df.PC1, df.PC2)
```

```python
source = ColumnDataSource(df)

hover = HoverTool(
        tooltips=[
            ("Name", "@Name"),
            ("Height", "@Height"),
            ("Weight", "@Weight"),
            ("Age", "@Age"),
            ("Position", "@Position"),
            ("Team", "Team")
        ]
    )
```

```python
p = figure(plot_width=800, plot_height=800, tools=[hover, 'wheel_zoom', 'pan'], title="MLB Players PCA")

p.circle('PC1', 'PC2', source=source, size=10, fill_color='blue')
p.xaxis.axis_label = 'PC 1'
p.yaxis.axis_label = 'PC 2'
show(p)
```
